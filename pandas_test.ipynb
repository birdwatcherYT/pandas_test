```python
import numpy as np
import json
import pandas as pd
```

```python
# csvロード
df=pd.read_csv("data.csv")
df
```

```python
# index設定
df=df.set_index("id")
df
```

```python
# mapによる置換
df["sex"]=df["sex"].map({"F":0, "M":1})
df
```

```python
# 日付の処理
df["time"]=pd.to_datetime(df["time"])
df
```

```python
# mapで各要素へ任意の処理
df['day']=df['time'].map(lambda x: x.day)
df
```

```python
df['weekday']=df['time'].map(lambda x: x.weekday())
df
```

```python
# NaNの数集計
df.isna().sum()
```

```python
# NaNを平均で埋める
df["age"]=df["age"].fillna(df["age"].mean())
df
```

```python
# 要素アクセス
df.loc[0, "name"]="Ａ"
df
```

```python
# 文字列の標準化
df["name"]=df["name"].str.normalize("NFKC")
df
```

```python
# 列名の変更
df.columns=([c.upper() for c in df.columns])
df
```

```python
# onehot特徴の追加
onehot = lambda target, num: np.eye(num,dtype=int)[target]
df[["weekday_"+str(i) for i in range(7)]] = pd.DataFrame(onehot(df["WEEKDAY"], 7))
df
```

```python
# データフレーム同士の連結
df=pd.concat([df, pd.DataFrame([[1,2],[3,4],[5,6],[7,8]], columns=["x", "y"])], axis=1)
df
```

```python
# 列追加
df["z"]=[1,2,3,4]
df
```

```python
# 条件による抜き出し
df[df["SEX"]==1]
```

```python
# 転置
df.T
```

```python
# データフレームに対する任意の処理
df[["xx","yy"]] = df[["x","y"]].applymap(lambda x: x*x)
df
```

```python
# 列削除1
del df["x"]
df
```

```python
# 列削除2
df = df[[c for c in df.columns if not c=="y"]]
df
```

```python
# 辞書へ変換
df.to_dict()
```

```python
# 集計処理
df.groupby(["SEX"]).mean()
```

```python
# 種類の確認
df["SEX"].unique()
```

```python
# 左結合
# week=pd.DataFrame([[0,"月"],[1,"火"],[2,"水"]],columns=["番号","曜日"])
# df=pd.merge(df, week, left_on="WEEKDAY", right_on="番号", how="left")
week=pd.DataFrame([[0,"月"],[1,"火"],[2,"水"]],columns=["WEEKDAY","曜日"])
df=pd.merge(df, week, on="WEEKDAY", how="left")
df
```

```python
# csvで保存
df.to_csv("__result__.csv",index=False)
```

```python
# jsonで保存
df.to_json("__result__.json")
```
